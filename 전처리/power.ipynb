```python
import pandas as pd 

df = pd.read_csv(r'C:\Users\82107\Desktop\4일 프로젝트 코드\pc_assembly\가공데이터\power.csv')
df.info()
```

```python
df['제조사'] = df['제조사'].str.replace(r'\( 제조사 웹사이트 바로가기 \)','',regex=True)
#값이없는 컬럼제거
df = df.drop(['DC 출력','[커넥터]','부가기능','Unnamed: 16','내부설계','보호회로','KC인증','출력','[변경사항]'],axis=1)
#UPS 관련 제거
df = df[df['제품 분류'] != 'UPS']
#값이없는 컬럼제거
df = df.dropna(axis=1, how='all')
```
