```python
#구형/폐기 소캣
df = df.drop(columns=['LGA775','소켓939','소켓754','소켓940','AM1'],axis=1)
#구형/특수 전원
df = df.drop(columns=['팬 24V','5V팬 3핀','IDE전원 4핀'],axis=1)
#필요없는정보 
df = df.drop(columns=['KC인증','지지대 방향','높이 조절범위','Unnamed: 8'],axis=1)
#값이없는 컬럼제거
df = df.drop(columns=['제품용도','기타 호환','부가기능','LED시스템','리모콘 지원','노트북용','구성품','[쿨링팬]','AMD 소켓','인텔 소켓'],axis=1)



```

```python
water_cool = df['냉각 방식'] == "수랭"
air_cool = df['냉각 방식'] == '공랭'
unknown_cool = (df['냉각 방식'] != '수랭') & (df['냉각 방식'] != '공랭')

water = df[water_cool].copy()
air = df[air_cool].copy()
unknown = df[unknown_cool].copy()
```

```python
#unknown[unknown['제품 종류'].isna()]
water['M.2 22110'].unique()
```

```python
unknown = unknown[~unknown['제품 종류']. isin (['튜닝 용품','Nan','가이드','써멀컴파운드','써멀퍼티','써멀패드','써멀 페이스트 가드','써멀컴파운드(그리스)',"써멀유형"])]
```

```python
#값이없는 컬럼제거
unknown.dropna(axis=1, how='all', inplace=True)
unknown=unknown.drop(['작동전압','튜닝용품 상세'],axis=1)
```

```python
#값이 두개이하인 컬럼 제거 
unknown.dropna(axis=1, thresh=3, inplace=True)
```

```python
#값이 없는 컬럼 제거 
water.dropna(axis=1, how='all', inplace=True)
air.dropna(axis=1, how='all', inplace=True)
```

```python
water=water.drop(['작동전압','써멀유형','써멀컴파운드'],axis=1)
air=air.drop(['작동전압','써멀유형','써멀컴파운드'],axis=1)
```

```python
unknown.to_csv('cooler_unknown.csv') #공랭 수랭 구분이없는것들
air.to_csv('cooler_air.csv') #공랭
water.to_csv('cooler_water.csv') #수랭
```
