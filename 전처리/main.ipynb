```python
import pandas as pd 

df = pd.read_csv(r'C:\Users\82107\Desktop\4일 프로젝트 코드\pc_assembly\가공데이터\mainboard.csv')
df
```

```python
import numpy as np

df['제조사'] = df['제조사'].str.replace(r'\( 제조사 웹사이트 바로가기 \)','',regex=True)
#데이터 없는 열 삭제
df = df.drop(['기본 사양','[메모리]','메모리 기술','[확장슬롯]','PCIe버전','PCIe슬롯','저장장치 인터페이스','[저장장치]','M.2 연결','M.2 폼팩터','RAID 지원','그래픽 출력','후면단자','네트워크','오디오','USB/팬 헤더','I/O 헤더','특징','KC인증'],axis=1)
df = df.drop(['Unnamed: 13','메모리 폼팩터','무선랜 종류','LED 위치'],axis=1)
#구형 기술 + 산업용 포트 + 개인용 pc에 불필요하다 생각하는 컬럼 삭제 
df = df.drop(['시리얼포트','LPT포트 헤더','침입감지 헤더','COM포트 헤더','PS/2'],axis=1)
#임베디드용 제거
df = df[df['제품 분류'] != '임베디드']
#산업용 제거 
df = df.drop(['GPIO','eMMC(내장)','mini-PCIe(mPCIe)','SFP+ 단자'],axis=1)
#내부 테스트용 컬럼 제거
df = df.drop(['안전부팅','OC프로파일','OC','BIOS선택'],axis=1)
# 같은 내용의 컬럼 통합 
df['USB3.0 Type-C 헤더'] = df['USB3.0 C타입 헤더'].fillna(df['USB3.0 Type C 헤더'].replace(0, np.nan)).fillna(0).astype(int)
df = df.drop(['USB3.0 C타입 헤더','USB3.0 Type C 헤더'],axis=1)
df['a'] = df['USB 2.0'].fillna(df['USB2.0'].replace(0, np.nan)).fillna(0).astype(int)
df = df.drop(['USB 2.0','USB2.0'],axis=1)
df['USB 2.0'] = df['a']

```
