```python
df['제조사'] = df['제조사'].str.replace(r'\( 제조사 웹사이트 바로가기 \)','',regex=True)
#데이터가없는 컬럼제거
df = df.drop(['칩셋 사양','메모리 사양','그래픽 출력','지원정보','전력 관련','냉각방식','제품 외형','부가기능','LED 위치','구성품','KC인증','Unnamed: 10','출력단자','전체 칩셋'],axis=1)
df = df.fillna('정보 없음')
#중복컬럼제거(다른컬럼으로 대체가능한것들)
df = df.drop(['제품 시리즈','NVIDIA 칩셋','AMD 칩셋','인텔 칩셋','수냉 쿨링','HDCP 지원'],axis=1)
#그래픽카드가 아닌 제품 제거 
df = df[df['그래픽카드 구성'] == '정보없음']
df = df[df['파워 구성'] == '정보없음']
df = df.drop(['그래픽카드 구성','파워 구성'],axis=1)
#그래픽
```
