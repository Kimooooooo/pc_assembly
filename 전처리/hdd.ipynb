```python
import pandas as pd 

df = pd.read_csv(r'C:\Users\82107\Desktop\4일 프로젝트 코드\pc_assembly\가공데이터\ssd.csv')
df.info()
```

```python
df['제조사'] = df['제조사'].str.replace(r'\( 제조사 웹사이트 바로가기 \)','',regex=True)
#값이없는 컬럼제거
df = df.drop(['부가기능','내구성/소음','A/S 정보','KC인증','Unnamed: 12'],axis=1)
#의미가 비슷한 컬럼 통합 
suri = ['5년','3년','1년','2년']
df['as기간'] = df[suri].apply(lambda row : ', '.join([col for col, val in row.items() if val == '○']), axis=1)
df = df.drop(suri,axis=1)
df['as기간'] = df['as기간'].replace('','정보없음') 
```

```python
df.to_csv('hdd.csv')
```
