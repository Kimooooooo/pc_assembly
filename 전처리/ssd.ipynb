```python
import pandas as pd

df = pd.read_csv(r'C:\Users\82107\Desktop\4일 프로젝트 코드\pc_assembly\원시데이터\추출폴더\ssd.csv')
df.info()
```

```python
df['제조사'] = df['제조사'].str.replace(r'\( 제조사 웹사이트 바로가기 \)','',regex=True)
#값이없는 컬럼제거
df = df.drop(['[기본사양]','[성능]','[환경특성]','Unnamed: 26','지원기능','부가기능','소프트웨어 지원','A/S기간','쿨링 정보','[크기/무게]','KC인증','색상','기타'],axis=1)
#중요도가낮고 값이 적은 컬럼제거  
df = df.drop(['데이터 복구 3년','데이터 복구 1년','맥북 업그레이드용','공정'],axis=1)
#의미가 비슷한 컬럼 통합 
suri = ['5년','3년','1년','7년','2년','10년']
color = ['블랙','옐로우','그린','블루','퍼플','화이트','그레이','실버','오렌지','레드','핑크']
df['as기간'] = df[suri].apply(lambda row : ', '.join([col for col, val in row.items() if val == '○']), axis=1) 
df['색상'] = df[color].apply(lambda row : ', '.join([col for col, val in row.items() if val == '○']), axis=1) 
df = df.drop(suri,axis=1)
df = df.drop(color,axis=1)
df= df.fillna('정보없음')
df['as기간'] = df['as기간'].replace('','정보없음')
df['색상'] = df['색상'].replace('','정보없음')
```

```python
df.to_csv('ssd.csv')
```
